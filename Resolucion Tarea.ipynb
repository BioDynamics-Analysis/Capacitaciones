```python
import pandas as pd
```

```python
df = pd.read_csv('data/data1.csv')
df.head()
```

```python
#1
df['food/shrimp (g)'] = df['food(kg)']/df['density(shrimp)'] * 1000
df.head()
```

```python
#2
df.groupby('days')['weight(g)'].agg(['min','max','mean']).head()
```

```python
#3 
su_prom = df['surviving'].mean()
df[ df['surviving']>su_prom ]
```

```python
#4
df['week2'] = df['week'].astype(int)
df.groupby('week2')['growth'].agg(['min','max'])
```

```python
#5
df_excel = pd.ExcelFile('data/Tabla Saturacion DO.xlsx').parse(0)
df_excel.dropna(axis=1, how='all')
```
