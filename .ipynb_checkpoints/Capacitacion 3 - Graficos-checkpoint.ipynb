```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/data1.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.head()
```

```python
df.loc['2020-07-15':].groupby('pool_id')['FCA'].plot(figsize=(15,5))
plt.title('Supervivencia por piscina')
plt.ylabel('Supervivencia')
plt.xlabel('Fecha')
plt.legend(title='ID pool')
plt.grid(alpha=.4)
# plt.yticks(np.arange(0,21, 4))
plt.xticks(rotation=0)
plt.savefig('Supervivencia por piscina.png')
plt.show()
```

```python
df.groupby('pool_id')['FCA'].last().plot(kind='bar')
```

```python
sns.heatmap(df[['surviving','density(shrimp)','density(shrimp/ha)','weight(g)']].corr(), annot=True, fmt='.1g')
plt.title('Correlaciones')
plt.show()
```
