```python
def DaysToWeek(x):
    return x/7

df['week2'] = df['week']/7
df['week3'] = df['week'].apply(DaysToWeek)
df['week4'] = df['week'].apply(lambda x: x/7)

#Crear columnas que me digan que piscinas son buenas
def catPool(x):
    if x<1.5:
        return 'Buena'
    return 'Mala'

df['cat'] = df['FCA'].apply(catPool)
df['cat1'] = df['FCA'].apply(lambda x: 'Buena' if x<1.5 else 'Mala')

df.head()
```

```python
#Filtrado con apply
df[ df['FCA'].apply(lambda x: x<1.5) ]
df[ df['FCA']<1.5 ]
```

```python
#apply con varias columnas
def catPool2(x):
    fca = x['FCA']
    su = x['surviving']
    
    if fca<1.2 and su>50:
        return 'Buena'
    return 'Mala'

df['cat4'] = df[['FCA','surviving']].apply(catPool2, axis=1) #1fila, 0 col
df['cat5'] = df[['FCA','surviving']].apply(lambda x: 'Buena' if x['FCA']<1.2 and x['surviving']>50 else 'Mala', axis=1) 
df.head()
```

```python
#Agupaciones
df.groupby('pool_id')['surviving'].min()
df3 = df.groupby('pool_id')[['surviving','weight(g)']].agg(['min','mean'])
```

```python
#Peso promedio por psicina y por semana
df['weekEntera'] = df['days']//7
df2 = df.groupby(['pool_id','weekEntera'])['weight(g)'].mean()
```
